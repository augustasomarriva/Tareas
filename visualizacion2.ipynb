```python
import matplotlib.pyplot as plt


anios = ['2019', '2020', '2021', '2022', "2023"] #Años
grupo_A = [30, 29, 17, 18, 19]  # Porcentajes para el grupo A en cada año
grupo_B = [26, 22, 14, 14, 15]  # Porcentajes para el grupo B en cada año
grupo_C = [25, 22, 29, 30, 31]  # Porcentajes para el grupo C en cada año
grupo_D = [19, 27, 40, 39, 36]  # Porcentajes para el grupo D en cada año


# Crear el gráfico de barras apiladas
plt.figure(figsize=(10, 5) )
 # Tamaño opcional para la figura (ancho, alto)

# Barras para cada grupo de personas
plt.bar(anios, grupo_A, label='C1', color='MediumSlateBlue')
plt.bar(anios, grupo_B, bottom=grupo_A, label='C2', color='DarkSlateBlue')
plt.bar(anios, grupo_C, bottom=[grupo_A[j] + grupo_B[j] for j in range(len(grupo_A))], label='C3', color='SlateBlue')
plt.bar(anios, grupo_D, bottom=[grupo_A[j] + grupo_B[j] + grupo_C[j] for j in range(len(grupo_A))], label='D', color='Indigo')


# Personalizar el gráfico
plt.xlabel('Grupo socioeconómico')   # Etiqueta del eje x
plt.ylabel('Porcentaje')      # Etiqueta del eje y
plt.title('Grupos socioeconómicos consumidores online en Chile')  # Título del gráfico
plt.ylim(0, 100)  # Ajustar el rango del eje y de 0 a 100%
plt.legend()  # Mostrar leyenda

# Mostrar el gráfico
plt.show()


```
